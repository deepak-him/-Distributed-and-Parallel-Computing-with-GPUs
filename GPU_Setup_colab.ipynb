```python
#Verify GPU Availability
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
```

torch.cuda.is_available() should return True.
torch.cuda.get_device_name(0) will show the name of your GPU.

```python
#list out all GPUs avaialble
import torch
if torch.cuda.is_available():
    # Get the number of GPUs available
    gpu_count = torch.cuda.device_count()
    print(f"Number of GPUs available: {gpu_count}")

    # List details of each GPU
    for i in range(gpu_count):
        gpu_name = torch.cuda.get_device_name(i)
        print(f"GPU : {gpu_name}")
else:
    print("CUDA is not available. Only CPU support.")
```

Set Device for Computation: To ensure PyTorch uses GPU for computations, you can explicitly move tensors and models to the GPU

```python
#Example
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
#  moving a tensor to the GPU
tensor = torch.tensor([1, 2, 3])
tensor = tensor.to(device)
```
